```python
def X_to_Xntime(origin, ntime_in,target):
    nsample = len(origin) - ntime_in + 1
    X_train = np.empty((nsample, ntime_in, 7))  # 배열 크기 미리 할당
    y_train = np.array([0 for i in range(nsample)])  # 배열 크기 미리 할당
    
    for i in range(nsample):
        new = np.copy(origin)
        X_train[i] = new[i:i+ntime_in] #.reshape(-1, 1)  # Numpy 배열에 직접 접근

    for i in range(nsample):
        y_train[i] = target

    return X_train, y_train
```

```python
anormal1__1X ,anormal1_1y = X_to_Xntime(anormal1_1,10,1)
anormal1__2X ,anormal1_2y = X_to_Xntime(anormal1_2,10,1)
anormal2__1X ,anormal2_1y = X_to_Xntime(anormal2_1,10,1)
anormal3__1X ,anormal3_1y = X_to_Xntime(anormal3_1,10,1)
anormal4__1X ,anormal4_1y = X_to_Xntime(anormal4_1,10,1)
anormal5__1X ,anormal5_1y = X_to_Xntime(anormal5_1,10,1)

normal1_X ,normal1_y = X_to_Xntime(normal1,10,0)
normal2_X ,normal2_y = X_to_Xntime(normal2,10,0)
normal3_X ,normal3_y = X_to_Xntime(normal3,10,0)
normal4_X ,normal4_y = X_to_Xntime(normal4,10,0)
normal5_X ,normal5_y = X_to_Xntime(normal5,10,0)
normal6_X ,normal6_y = X_to_Xntime(normal6,10,0)
normal7_X ,normal7_y = X_to_Xntime(normal7,10,0)
normal8_X ,normal8_y = X_to_Xntime(normal8,10,0)
```

```python
anormal_X = np.concatenate((anormal1__1X, anormal1__2X,anormal2__1X,anormal3__1X,anormal4__1X,anormal5__1X), axis=0)
normal_X = np.concatenate((normal1_X, normal2_X, normal3_X, normal4_X, normal5_X, normal6_X, normal7_X, normal8_X), axis=0)
anormal_y = np.concatenate((anormal1_1y, anormal1_2y,anormal2_1y,anormal3_1y,anormal4_1y,anormal5_1y), axis=0)
normal_y = np.concatenate((normal1_y, normal2_y, normal3_y, normal4_y, normal5_y, normal6_y, normal7_y, normal8_y), axis=0)

X = np.concatenate((anormal_X,normal_X), axis=0)
y = np.concatenate((anormal_y,normal_y), axis=0)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
import tensorflow as tf
# Sequential 모델 생성
model = tf.keras.Sequential()

# Sequential 모델 생성
model = tf.keras.Sequential()

# 입력 레이어 (입력 데이터의 shape를 지정)
model.add(tf.keras.layers.InputLayer(input_shape=(10, 7)))

# Flatten 레이어 (2D 배열을 1D 배열로 평탄화)
model.add(tf.keras.layers.Flatten())

#model.add(tf.keras.layers.LSTM(units=32,input_shape=(10, 7)))
# 은닉 레이어
model.add(tf.keras.layers.Dense(units=128, activation='relu'))
model.add(tf.keras.layers.Dense(units=16, activation='relu'))
model.add(tf.keras.layers.Dense(units=128, activation='relu'))
# 출력 레이어 (예측 값)
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# 모델 요약 출력
model.summary()

# 모델 컴파일
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 모델 훈련
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
```

```python
# 학습 결과 그래프로 나타내기
plt.figure(figsize=(12, 6))

# 학습 및 검증 손실 그래프
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# 학습 및 검증 정확도 그래프
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```
